```python
# %matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import skimage
from skimage import io
from skimage.viewer import ImageViewer
```

```python
def get_neighborhood(im, x_, y_, d_size):
    size = int(d_size / 2)
    if d_size % 2 == 0:
        return im[x_ - size + 1: x_ + size + 1, 
                  y_ - size + 1: y_ + size + 1, 
                  0]
    return im[x_ - size : x_ + size + 1, 
                  y_ - size : y_ + size + 1, 
                  0]

def apply_mask(im, mask): #aplico as mascaras mask e devolvo uma imagem nova com a mascara aplicada
    mask = np.array(mask)
    x_max, y_max, _ = im.shape #pego altura e largura do im.shape
    neigh_size, _ = mask.shape #
    
    new_im = im.copy() #nova imagem para aplicar a mascara
    for x in range(1, x_max - 1):
        for y in range(1, y_max - 1):
            neighs = np.array(get_neighborhood(im, x, y, neigh_size))   # do pixel x,y da img original, com o tamanho dos vizinhos
            applied_matrix = neighs * mask #calculo a mascara com os vizinhos
            summed = applied_matrix.flatten().sum() #soma o total e aplica a mascara
            new_im[x][y] = summed
    return new_im
```

```python
def calculate_threshold(im): # Calculando valor de threshold global com base no histograma
    # Assumindo um histograma bimodal, com 
    hist_arr = np.repeat(0, 256)
    for px in im[:,:,0].flatten():
        hist_arr[px] += 1

    min_val = 256
    min_i = 0
    for i, val in list(enumerate(hist_arr))[40:230]:
        if min_val < val:
            min_val = val #valor
            min_i = i #posição
    return min_i, min_val, hist_arr

def adaptative_global_threshold(im):
    t, _, _ = calculate_threshold(im)
    return global_threshold(im, t)
```

```python
im = io.imread('images/tyrion.jpg')
im = grayify(im)

f, ax = plt.subplots(2, 1, figsize=(16, 16), dpi=80)
ax[0].axis('off')
ax[1].axis('off')

im_thresholded = adaptative_global_threshold(im)

ax[0].imshow(im)
ax[1].imshow(im_thresholded)
```
